```python
# Get a dictionary with column names as keys and lists of unique values as values
unique_values = {col: df[col].unique().tolist() for col in df.columns}

# Print the result
for col, values in unique_values.items():
    print(f"{col}: {values}")
```

```python
# Select only columns with 'object' datatype
object_cols = df.select_dtypes(include=['object']).columns

# Get unique values for each 'object' column
unique_values = {col: df[col].unique().tolist() for col in object_cols}

# Print the result
for col, values in unique_values.items():
    print(f"{col}: {values}")
```

```python
value_counts = df['cuisine'].value_counts()

# Create a bar chart
plt.figure(figsize=(8, 5))
value_counts.plot(kind='bar', color='skyblue')
plt.title('Count of Meals per Cuisine')
plt.xlabel('cuisine')
plt.ylabel('Count')
plt.xticks(rotation=45)
plt.tight_layout()
plt.show()
```

```python
value_counts_is_healthy = df['is_healthy'].value_counts()

# Create a bar chart
plt.figure(figsize=(8, 5))
value_counts_is_healthy.plot(kind='bar', color='skyblue')
plt.title('Healthy vs non-healthy Meals')
plt.xlabel('is_healthy')
plt.ylabel('Count')
plt.xticks(rotation=45)
plt.tight_layout()
plt.show()
```

```python
plt.hist(df['calories'], bins=70, edgecolor='black', color='blue')
plt.xlabel('calories per meal')
plt.ylabel('number of meals')
plt.show()
```
