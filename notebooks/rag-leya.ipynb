## RAG

```python
from dotenv import load_dotenv
import warnings
from langchain_groq import ChatGroq
from langchain.prompts.prompt import PromptTemplate
```

```python
load_dotenv()               #load credentials
```

```python
warnings.filterwarnings("ignore")           #defining the llm (using the Groq)

llm = ChatGroq(
    model="llama-3.1-8b-instant", #"llama3-8b-8192": fast and small LLaMA 3 version
    temperature=0,                  #deterministic (not random) answers
    max_tokens=None,        #no limit unless default given by api
    timeout=None,           #not limited by timeout
    max_retries=2           #retry up to 2 times if the request fails
)
```

```python
from langchain.docstore.document import Document

#Convert each row to documents:
documents = [
    Document(page_content=row.to_json())
    for _, row in df.iterrows()    
]
```

```python
from langchain.embeddings import HuggingFaceEmbeddings

embedding_model = HuggingFaceEmbeddings(model_name="sentence-transformers/all-MiniLM-L6-v2")
#vectorize the text so it can be searched semantically
```

What the warning means:

This means that Hugging Face tried to use the hf_xet library (a tool for optimized data downloading via XetHub), but since it's not installed, it simply defaulted to regular HTTP download â€” which works just fine, just slightly slower.

```python
from langchain.vectorstores import FAISS
#FAISS is a vectore store library for quick similarity search
#storing in a Vector Store
db = FAISS.from_documents(documents, embedding_model)
```

```python
retriever = db.as_retriever()
#db stores docs as embeddings
# .as_retriever() wraps the DB for use in a standard LangChain chains
#retriever fetches relevant dochs for a qeury
```

```python
from langchain.chains import RetrievalQA    
#seting up a RAG Chain
qa_chain = RetrievalQA.from_chain_type(
    llm=llm,
    retriever=retriever,
    return_source_documents=True
)
```

```python
query = "Can you summarize my dataset? Please do it thouroghly"
result = qa_chain(query)

print(result['result'])
```

```python
query = "Based on the dataset, please make a low-carb recipe for dinner for two persons."
result = qa_chain(query)

print(result['result'])
```
